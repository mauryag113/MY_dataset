```python
import os
from ultralytics import YOLO

# Set the path to the data.yaml file you created
data_path = '/content/drive/MyDrive/Object Detection.v1i.yolov8/data.yaml'

# Load YOLOv8 model (this will load the pre-trained YOLOv8 model for transfer learning)
model = YOLO('yolov8n.pt')  # You can choose different sizes like yolov8s.pt, yolov8m.pt, yolov8l.pt, or yolov8x.pt

# Train the model
model.train(
    data=data_path,       # Path to your dataset config file
    epochs=100,           # Number of epochs to train
    batch=16,             # Batch size
    imgsz=640,            # Image size (can be adjusted depending on GPU memory)
    lr0=0.01,             # Learning rate (can be adjusted)
    lrf=0.1,              # Learning rate final
    momentum=0.937,       # Momentum
    weight_decay=0.0005,  # Weight decay
    save_period=1,        # Save checkpoint every epoch
    project='object_detection_project',  # Project directory where the results will be saved
    name='yolov8_training',  # Experiment name
    exist_ok=True         # Overwrite existing project folder if necessary
)

# Evaluate the trained model
metrics = model.val()
print("Validation Metrics:", metrics)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns
import os
import cv2
from ultralytics import YOLO

# Load the trained YOLOv8 model
model = YOLO('/content/yolov8n.pt')  # Path to your best-trained weights

# Predict on the test images
results = model.predict(
    source='/content/drive/MyDrive/Object Detection.v1i.yolov8/test/images',  # Path to your test images
    conf=0.5                        # Confidence threshold
)

# Ground truth labels and predicted labels for confusion matrix
y_true = []  # Ground truth labels (for all test images)
y_pred = []  # Predicted labels (for all test images)

# Define a function to extract ground truth labels
def extract_ground_truth(image_path):
    label_path = image_path.replace('images', 'labels').replace('.jpg', '.txt')  # Adjust if using other formats
    ground_truth = []
    if os.path.exists(label_path):
        with open(label_path, 'r') as file:
            for line in file.readlines():
                class_id = int(line.strip().split()[0])
                ground_truth.append(class_id)
    return ground_truth

# Iterate over results to extract predictions and compare with ground truth
for result in results:
    # Extract the predicted class labels (last column)
    predictions = result.boxes.cls.cpu().numpy()  # Get predicted class labels

    # Print the predictions for debugging
    print(f"Predictions for {result.path}: {predictions}")

    # Extract ground truth from corresponding annotation file
    ground_truth = extract_ground_truth(result.path)  # Use result.path to get the image path

    # Print the ground truth for debugging
    print(f"Ground truth for {result.path}: {ground_truth}")

    # Update y_true and y_pred with actual values if there are predictions
    if len(predictions) > 0:
        y_true.extend(ground_truth)
        y_pred.extend(predictions)

# Check if y_true and y_pred are of the same length
print(f"Length of y_true: {len(y_true)}")
print(f"Length of y_pred: {len(y_pred)}")

# Only generate confusion matrix if the lengths match
if len(y_true) == len(y_pred):
    # Generate confusion matrix
    cm = confusion_matrix(y_true, y_pred)

    # Plot the confusion matrix using seaborn
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 0', 'Class 1', 'Class 2'], yticklabels=['Class 0', 'Class 1', 'Class 2'])
    plt.title("Confusion Matrix")
    plt.xlabel("Predicted Labels")
    plt.ylabel("True Labels")
    plt.show()
else:
    print("Mismatch in the number of predictions and ground truth labels!")
```
